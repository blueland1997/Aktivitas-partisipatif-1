```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# a. Memanggil data
file_path = '/content/CAD alizadeh.xls'  # Path file yang diunggah
df = pd.read_excel(file_path)

# b. Menampilkan data
print("Data Awal:")
print(df.head())

# c. Preprocessing data: Memeriksa apakah ada data yang kosong
missing_values = df.isnull().sum()
print("\nJumlah nilai yang kosong di setiap kolom:")
print(missing_values)

# Mengisi data yang kosong dengan nilai median (atau metode lain jika diperlukan)
df.fillna(df.median(numeric_only=True), inplace=True)

# d. Mengubah semua kolom menjadi numerik
for col in df.columns:
    if df[col].dtype == 'object':  # Jika kolom bertipe string/objek
        df[col] = pd.factorize(df[col])[0]  # Mengubah kategori menjadi angka

print("\nData setelah konversi menjadi numerik:")
print(df.head())

# e. Normalisasi data
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("\nData setelah normalisasi:")
print(df_normalized.head())
```
